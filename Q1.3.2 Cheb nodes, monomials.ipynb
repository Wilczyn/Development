```python
import sympy as sy
import numpy as np
import matplotlib.pyplot as plt
import math
from math import exp

plt.style.use("ggplot")

# Define the variable and the function to approximate
#x = sy.Symbol('x')
#fe = sy.exp(1/x)
#runge = 1/(1+25*x**2)
#ramp = (x+abs(x))/2
```

```python
def plotFE():
    x1 = np.linspace(-1,1,800)
    y1 = np.exp(1/x1)

    
    n=4 # number of nodes
    
    x_train = np.zeros(n)
    i=1
    while i <= n:
        x_train[i-1] = math.cos((2*i-1)/(2*n)*math.pi)
        i += 1
    
    y_train = np.exp(1/x_train)
    #Estimate coefficents of polynomial fit
    CoefPolynomial3_FE = np.polyfit(x_train, y_train, 3)
    CoefPolynomial5_FE = np.polyfit(x_train, y_train, 5)
    CoefPolynomial10_FE = np.polyfit(x_train, y_train, 10)
    
    Polynomial3 = np.poly1d(CoefPolynomial3_FE)
    Polynomial5 = np.poly1d(CoefPolynomial5_FE)
    Polynomial10 = np.poly1d(CoefPolynomial10_FE)
    
    plt.scatter(x_train,y_train, c= 'black', label='Interpolated points - Chebyshev nodes')
    plt.plot(x1,Polynomial3(x1),label='Cubic polynomial')
    plt.plot(x1,Polynomial5(x1),label='Polynomial of order 5')
    plt.plot(x1,Polynomial10(x1),label='Polynomial of order 10')
    
        # Plot the function
    plt.plot(x1,y1,label='exp(1/x)')
    
    plt.xlim([-1,1])
    plt.ylim([-10,40])
    plt.xlabel('x')
    plt.ylabel('y')
    plt.legend()
    plt.grid(True)
    plt.title('Exp(1/x), Chebyshev nodes, monomials')
    plt.rcParams["figure.figsize"] = (12,8)
    plt.show()

plotFE()
```
