```python
import numpy as np
import matplotlib.pyplot as plt

plt.style.use("ggplot")

#Define the CES function
alpha = 0.5
sigma = 1.01
K = np.linspace(0,10,1000)
H = np.linspace(0,10,1000)

l=1000
Y = [0] * l
for i in range(l):
    Y[i] = [0] * l

x_iso5 = []
y_iso5 = []
x_iso10 = []
y_iso10 = []
x_iso25 = []
y_iso25 = []
x_iso50 = []
y_iso50 = []
x_iso75 = []
y_iso75 = []
x_iso90 = []
y_iso90 = []
x_iso95 = []
y_iso95 = []

for h in range(1000):
    for k in range(1000):
        Y[k][h] = ((1-alpha)*K[k]**((sigma-1)/sigma)+alpha*H[h]**((sigma-1)/sigma))**(sigma/(sigma-1)) #Exact values
        if Y[k][h] >= 0.48 and Y[k][h] <= 0.52: 
            x_iso5.append(K[k])
            y_iso5.append(H[h])
        if Y[k][h] >= 0.98 and Y[k][h] <= 1.02: 
            x_iso10.append(K[k])
            y_iso10.append(H[h])
        if Y[k][h] >= 2.48 and Y[k][h] <= 2.52: 
            x_iso25.append(K[k])
            y_iso25.append(H[h])
        if Y[k][h] >= 4.98 and Y[k][h] <= 5.02: 
            x_iso50.append(K[k])
            y_iso50.append(H[h])
        if Y[k][h] >= 7.48 and Y[k][h] <= 7.52: 
            x_iso75.append(K[k])
            y_iso75.append(H[h])
        if Y[k][h] >= 8.98 and Y[k][h] <= 9.02: 
            x_iso90.append(K[k])
            y_iso90.append(H[h])
        if Y[k][h] >= 9.48 and Y[k][h] <= 9.52: 
            x_iso95.append(K[k])
            y_iso95.append(H[h])
            
    
plt.plot(x_iso5, y_iso5, label='isoquants of the percentile 5')
plt.plot(x_iso10, y_iso10, label='isoquants of the percentile 10')
plt.plot(x_iso25, y_iso25, label='isoquants of the percentile 25')
plt.plot(x_iso50, y_iso50, label='isoquants of the percentile 50')
plt.plot(x_iso75, y_iso75, label='isoquants of the percentile 75')
plt.plot(x_iso90, y_iso90, label='isoquants of the percentile 90')
plt.plot(x_iso95, y_iso95, label='isoquants of the percentile 95')
plt.rcParams["figure.figsize"] = (12,12)
plt.legend()
plt.title('Isoquants of CES function, sigma = ' + str(sigma))
```
